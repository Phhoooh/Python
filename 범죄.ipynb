```python
crime_seoul1 = crime_seoul.dropna(axis=0, how='any')
```

```python
station_name = []
for name in crime_seoul1['구분']:
    station_name.append('서울' + str(name[:] + '경찰서'))

station_name
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name,language = 'ko')
    station_address.append(tmp[0].get("formatted_address"))
    
    tmp_loc =tmp[0].get("geometry")
    station_lat.append(tmp_loc['location']['lat'])
    
    station_lng.append(tmp_loc['location']['lng'])
    print(name + ' --> ' + tmp[0].get("formatted_address"))
```

```python
gu_name = []

for name in station_address:
    tmp = name.split()
    
    tmp_gu = [gu for gu in tmp if gu[-1] == '구'][0]
    
    gu_name.append(tmp_gu)
    
crime_seoul1['구별'] = gu_name
crime_seoul1.head()
```

```python
crime_anal = pd.pivot_table(crime_seoul1, index='구별', columns=['죄종','발생검거'], values='건수',  aggfunc = np.sum)
crime_anal.head()
```

```python
crime_anal['강간검거율'] = crime_anal['강간']['검거'] / crime_anal['강간']['발생']
crime_anal['강도검거율'] = crime_anal['강도']['검거'] / crime_anal['강도']['발생']
crime_anal['살인검거율'] = crime_anal['살인']['검거'] / crime_anal['살인']['발생']
crime_anal['절도검거율'] = crime_anal['절도']['검거'] / crime_anal['절도']['발생']
crime_anal['폭력검거율'] = crime_anal['폭력']['검거'] / crime_anal['폭력']['발생']
```
